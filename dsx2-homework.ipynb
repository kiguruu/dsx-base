```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# 色名の取り出し
cnames = mpl.colors.cnames

# そこから上位20個を取り出し
names = list(cnames.keys())[:20]
colors = [cnames[name] for name in names]

# 棒を用意して
x = np.arange(1, 21)
height = np.repeat(1, 20)

# 表示
plt.figure(figsize =(10, 6))
# ラベルを45度傾けて、右端を軸に合わせる
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.bar(x, height, color=colors, tick_label=names)
```
