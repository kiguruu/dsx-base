```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 2, 3], label='firstline')
ax.plot([1, 2, 3, 4], [5, 7, 11, 4], label='secondline')
plt.legend();
```

```python
x1 = np.linspace(0, 2, 100)  # Sample data.
y1 = x1**2
x2 = np.linspace(0, 5*np.pi, 100)
y2 = np.sin(x2)
fig, ax = plt.subplots()
ax.plot(y1, label="straight")
ax.plot(y2, label="curve")
plt.legend();
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 2, 3], label='line', color="red", 
linewidth=2, linestyle="dashed")
plt.legend();
```

```python
from pyvis.network import Network
G = Network()
G.add_nodes([1, 2], label=['Alex', 'Carthy'])
G.add_nodes([3, 4, 5, 6], 
              label=['Michael', 'Ben', 'Oliver', 'Olivia'],
              color=['#3da831', '#9a31a8', '#3155a8', '#eb4034'])
G.add_edges([(1, 5, 2), (2, 5, 5), (3, 4, 2), (1, 6), (2, 6), (3, 5)])
G.show('edges_with_weights.html', notebook=False)
```
